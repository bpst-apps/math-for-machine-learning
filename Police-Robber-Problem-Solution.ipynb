### Police-Robber Problem

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define time variable
t = np.linspace(0, 40, 1000) # start, finish, n points
```

Distance travelled by robber: ${d = 2.5t}$

```python
# define linear equation for robber
distance_robber = 2.5 * t
```

Distance travelled by police: ${d = 3(t-5)}$

```python
# define linear equation for police
distance_police = 3 * (t - 5)
```

```python
# plotting the equations
fig, ax = plt.subplots()

plt.title('Robber Police Problem')
plt.xlabel('time (minutes)')
plt.ylabel('distance (km)')

ax.set_xlim([0, 40])
ax.set_ylim([0, 100])

# plotting both linear equations
ax.plot(t, distance_robber, c='green')
ax.plot(t, distance_police, c='brown')

# time = 30 minutes, distance = 75 Km
plt.axvline(x=30, color='purple', linestyle='--')
plt.axhline(y=75, color='purple', linestyle='--')

plt.show()
```
